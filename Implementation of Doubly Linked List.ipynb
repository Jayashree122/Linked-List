```python
#Implementation of Doubly Linked List
class Node:
  def __init__(self,data):
    self.prev=None
    self.data=data
    self.next=None
class DoublyLinkedList:
  def __init__(self):
    self.head=None
  def insert_at_beg(self,newnode):
    if self.head==None:
      self.head=newnode
    else:
      newnode.next=self.head
      self.head.prev=newnode
      self.head=newnode
  def insert_at_end(self,newnode):
    if self.head==None:
      self.head=newnode
    else:
      temp=self.head
      while(temp.next!=None):
        temp=temp.next
      temp.next=newnode;
      newnode.prev=temp;
      newnode.next=None
  def insert_at_pos(self,pos,newnode):
    if self.head:
      temp=self.head
      for i in range(0,pos-2):
        temp=temp.next
      newnode.next=temp.next
      temp.next.prev=newnode
      temp.next=newnode
      newnode.prev=temp
    else:
      self.head=None
  def delete_at_beg(self):
    if self.head==None:
      print("Empty")
    else:
      temp=self.head
      self.head=temp.next
      self.head.prev=None
      temp.next=None
  def delete_at_pos(self,pos):
    if self.head:
      temp=self.head
      for i in range(1,pos-2):
        temp=temp.next
      temp2=temp.next
      temp.next=temp2.next
      temp2.next.prev=temp
      temp2.next=None
      temp2.prev=None
    else:
      print("Empty")
  def delete_at_end(self):
    if self.head==None:
      print("Empty")
    else:
      temp=self.head
      while(temp.next.next!=None):
          temp=temp.next
      temp.next=None
  def traversal(self):
    if self.head==None:
      print("Empty")
    else:
      temp=self.head
      while(temp!=None):
        print(temp.data,end=' ')
        temp=temp.next
      print()
```
