```python
#IMPLEMENTATION OF CIRCULAR SINGLY LINKED LIST
class node:
  def __init__(self,val):
    self.data=val
    self.next=None
class Csll:
  def __init__(self):
    self.head=None
    self.tail=None
  def ins_at_beg(self,newnode):
    if self.head:
      newnode.next=self.head
      self.head=newnode
      self.tail.next=self.head
    else:
      self.head=self.tail=newnode
      self.tail.next=self.head
  def ins_at_end(self,newnode):
    if self.head:
      self.tail.next=newnode
      self.tail=newnode
      self.tail.next=self.head
    else:
      self.head=self.tail=newnode
      self.tail.next=self.head
  def del_at_end(self):
  #o(n) time complexity in SLL
    if self.head:
      temp=self.head
    while temp.next!=self.tail:
      temp=temp.next
      temp.next=None
      self.tail=temp
      self.tail.next=self.head
    else:
      print('ll is empty')
  def del_at_beg(self):
    if self.head:
      self.head=self.head.next
      self.tail.next=self.head
    else:
      print('ll is empty')
  def print_ll(self):
    if self.head:
      temp=self.head
    while(1):
      print(temp.data,end=' ')
      temp=temp.next
      if temp==self.head:
        break
    print()
s=Csll()
s.ins_at_beg(node(10))
s.ins_at_beg(node(20))
s.ins_at_beg(node(30))
s.ins_at_beg(node(40))
s.ins_at_beg(node(50))
s.print_ll()
s.del_at_beg()
s.print_ll()
s.del_at_end()
s.print_ll()
s.ins_at_end(node(100))
s.ins_at_end(node(200))
s.print_ll()
```
