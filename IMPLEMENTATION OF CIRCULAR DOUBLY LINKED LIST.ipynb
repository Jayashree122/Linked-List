```python
#IMPLEMENTATION OF CIRCULAR DOUBLY LINKED LIST

class node:
  def __init__(self,ele):
    self.prev=None
    self.data=ele
    self.next=None
class Circular_Doubly_List:
  def __init__(self):
    self.head=None
    self.tail=None
  def ins_at_beg(self,newnode):
    if self.head==None:
      self.head=newnode
      self.tail=newnode
      self.tail.next=self.head
    else:
      newnode.next=self.head
      self.head.prev=newnode
      self.head=newnode
      self.tail.next=self.head
  def ins_at_end(self,newnode):
    if self.head==None:
      self.head=newnode
      self.tail=newnode
      self.tail.next=self.head
    else:
      newnode.prev=self.tail
      self.tail.next=newnode
      self.tail=newnode
      self.tail.next=self.head
  def del_at_beg(self):
    if self.head:
      self.head=self.head.next
      self.head.prev=None
      self.tail.next=self.head
    else:
      print('It is empty')
  def del_at_end(self):
    if self.tail:
      self.tail=self.tail.prev
      self.tail.next=None
      self.tail.next=self.head
    else:
      print('ll is empty')
  def display(self):
    if self.head:
      temp=self.head
      while 1:
        print(temp.data,end=' ')
        temp=temp.next
        if temp==self.head:
          break
      print()
```
