```python
#Implementation of Singly Linked List
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class SinglyLinkedList:
  def __init__(self):
    self.head=None
  def ins_at_beg(self,newnode):
    if self.head==None:
      self.head=newnode
    else:
      newnode.next=self.head
      self.head=newnode
  def ins_at_position(self,position,newnode):
    if self.head==None:
      self.head=newnode
    else:
      t=self.head
      for i in range(1,position-2):
        t=t.next
      newnode.next=t.next
      t.next=newnode
  def ins_at_end(self,newnode):
    if self.head==None:
      self.head=newnode
    else:
      t=self.head
      while(t.next!=None):
        t=t.next
      t.next=newnode
  def del_at_beg(self):
    if self.head==None:
      print("Empty")
    else:
      self.head=self.head.next
  def del_at_position(self,position):
    if self.head==None:
      print("Empty")
    else:
      t=self.head
      for i in range(1,position-2):
        t=t.next
      t.next=t.next.next
  def del_at_end(self):
    if self.head==None:
      print("Empty")
    else:
      t=self.head
      while(t.next.next!=None):
        t=t.next
      t.next=None
  def traversal(self):
    if self.head==None:
      print("Empty")
    else:
      t=self.head
      while t!=None:
        print(t.data,end=' ')
        t=t.next
      print()
```

```python
a=SinglyLinkedList()
a.ins_at_end(Node(2222))
a.ins_at_end(Node(2567))
a.ins_at_end(Node(4346))
a.ins_at_end(Node(30566))
a.traversal()
a.ins_at_position(3,Node(234))
a.ins_at_beg(Node(266))
a.traversal()
a.del_at_end()
a.del_at_beg()
a.del_at_position(2)
a.traversal()
```
